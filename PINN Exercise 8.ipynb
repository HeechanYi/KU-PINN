**Problem 1.**
Let $X$ and $Y$ be random variables with joint density function
$$
f(x,y) = 
\begin{cases}
4xy, & 0< x, y < 1 \\
0, & \mathrm{elsewhere}
\end{cases}
$$
Find the expected value of $Z = \sqrt{X^2 + Y^2}$.

----

**[Sol]**<br>
To calculate the expectation value of $Z$. We have to do the integral below:
$$ \int_0^1 \int_0^1 \sqrt{x^2 + y^2} \times f(x,y) \ dx dy $$
Consider integration with substitution,
$x^2 = t$ where $2xdx = dt$ and $y^2 = s$ where $2ydy = ds$.<br>
Then, the integral can be exchanged to
$$ \int_0^1 \int_0^1 \sqrt{t + s} \  dtds $$
Then,<br>
$
\begin{aligned}
\int_0^1 \int_0^1 \sqrt{t + s} \ dtds & = \int_0^1 \left[\frac{2}{3}(t+s)^{\frac{3}{2}} \right]^1_0 ds \\
& = \int_0^1 \left(\frac{2}{3}(1+s)^{\frac{3}{2}} - \frac{2}{3}s^{\frac{3}{2}} \right) ds\\
& = \left[ \frac{2}{3}\cdot\frac{2}{5}(1+s)^{\frac{5}{2}} \right]^1_0 - \left[\frac{2}{3}\cdot\frac{2}{5}s^{\frac{5}{2}} \right]^1_0 \\
& = \frac{4}{15} 2^{\frac{5}{2}} - \frac{4}{15} - \frac{4}{15} + 0 \\
& = \frac{8}{15} (2^{\frac{3}{2}} - 1 )
\end{aligned}
$

**Problem 2.**
The length of time, in minutes, for an airplane to obtain clearance for takeoff at a certain airport is given by a random variable $Y = 3X âˆ’ 2$, where $X$ has the density function
$$
f(x) = 
\begin{cases}
\frac{1}{4}e^{-x/4}, & x>0 \\
0, & \mathrm{elsewhere}
\end{cases}
$$
Find the mean and variance of the random variable $Y$.

----

**[Sol]**<br>
To calculate the mean, we calculate the expecation value of $X$ whih is
$$\bar{X} = E[X] = \int^\infin_0 x f(x) dx$$
Then, by integration by parts<br><br>
$
\begin{aligned}
\int^\infin_0 x f(x) dx & = \int^\infin_0 \frac{1}{4} x e^{-x/4} dx \\
& = \frac{1}{4} \left\{ \left[ x \cdot \left(-\frac{1}{4} e^{-x/4} \right)\right]^\infin_0 - \int^\infin_0 -\frac{1}{4} e^{-x/4} dx\right\} \\
& = \frac{1}{16} \int^\infin_0 e^{-x/4} dx \\
& = \frac{1}{16} \left[ -\frac{1}{4} e^{-x/4}\right]^\infin_0 \\
& = \frac{1}{64} 
\end{aligned}
$
<br><br>
Calculating the variance,
$$ \sigma^2 = E[(X - \bar{X})^2]$$
We can modify the variation like below:<br><br>
$
\begin{aligned}
E[(X - \bar{X})^2] & = E[X^2 - 2X\bar{X} + \bar{X}^2] \\
& = E[X^2] - 2\bar{X}E[X] + \bar{X}^2 \\
& = E[X^2] - \bar{X}^2
\end{aligned}
$
<br><br>
Since the expectation value of $X$ is the mean of the variable and also a constant, which can be get out of the calcualation of expectation vlaue.<br>
Then, we have to calculate only the expecatiotion value of the square of $X$ for solving the variance.<br><br>
$
\begin{aligned}
E[X^2] & = \int^\infin_0 \frac{1}{4} x^2 e^{-x/4} dx \\
& = \frac{1}{4} \left\{ \left[ x^2 \cdot \left(-\frac{1}{4} e^{-x/4} \right)\right]^\infin_0 - \int^\infin_0 -\frac{1}{2} x e^{-x/4} dx\right\} \\
& = \frac{1}{8} \int^\infin_0 e^{-x/4} dx \\
& = \frac{1}{8} \left[ -\frac{1}{4} e^{-x/4}\right]^\infin_0 \\
& = \frac{1}{32}
\end{aligned}
$

**Problem 3.**
Given a random variable $X$, with standard deviation $\sigma_x$, and a random varialbe $Y = a + bX$.<br>
Show that if $b<0$, the correlation coefficient $\rho_{XY} = -1$, and if $b>0$, $\rho_{XY} = 1$.

----

**[Sol]**<br>
The definition of the correlation coefficient is:
$$ \rho_{XY} \equiv \frac{\mathrm{Cov}(X,Y)}{\sigma_X \sigma_Y} = \frac{E[(X - \bar{X})(Y - \bar{Y})]}{\sigma_X \sigma_Y} $$ 
where $\bar{X}, \bar{Y}$ are the mean of the variables and $\sigma_X$,$\sigma_Y$ denotes the square root of the variance of each variables.<br>
Therefore, let $\bar{X} = \mu$ and $\sigma_X = \sigma$ then,<br><br>
$E[X] = \mu$ <br>
$E[Y] = E[a + bX] = a + b\mu$ <br>
$\sigma_Y = |b|\sigma$ since $\sigma_Y^2 = E[(Y - \bar{Y})^2] = b^2E[(X - \bar{X})^2] = b^2 \sigma^2$ <br>
Then, <br><br>
$
\begin{aligned}
\rho_{XY} & =  \frac{E[(X - \bar{X})(Y - \bar{Y})]}{\sigma_X \sigma_Y} \\
& = \frac{E[XY] - \bar{X}E[Y] - \bar{Y}E[X] +\bar{X}\bar{Y}}{|b|\sigma^2} \\
& = \frac{E[aX + bX^2] - \mu(a+b\mu)}{|b|\sigma^2} \\
& = \frac{a\mu + b(E[X^2] - \mu^2) + b\mu^2 - \mu(a+b\mu)}{{|b|\sigma^2}} \\
& = \frac{a\mu + b\sigma^2 + b\mu^2 - \mu(a+b\mu)}{|b|\sigma^2} \\
& = \frac{a\mu + b\sigma^2 + b\mu^2 - \mu(a+b\mu)}{|b|\sigma^2} \\
& = \frac{ b\sigma^2}{|b|\sigma^2} \\
& = \frac{b}{|b|} \\
\end{aligned}
$ <br>
So the correlation of $X$ and $Y$ depends on the sign of $b$ such that if $b<0$ then $\rho_{XY} = -1$ and if $b>0$ then $\rho_{XY} = 1$.

**Problem 4.**
Random variables $X$ and $Y$ follows a joint distribution
$$
f(x, y) =
\begin{cases}
2, & 0 < x \leq y < 1\\
0, & \mathrm{otherwise}
\end{cases}
$$
Determine the correlation coefficient between $X$ and $Y$.

----

**[Sol]**<br>
For solving the correlation coefficient, firstly we have to calculate the marginal probability distribuition of $X$ and $Y$.<br>
Then, by step by step <br><br>
**Step 1. marginal probabiliy distribution** <br>
$
\begin{aligned}
f_X(x) & = \int^1_x f(x, y) dy \\
& = \int^1_x 2 dy = \left[ 2y \right]^1_x \\
& = 2 - 2x
\end{aligned}
$ 
<br>
$
\begin{aligned}
f_Y(y) & = \int^y_0 f(x, y) dx \\
& = \int^y_0 2 dx = \left[ 2x \right]^y_0 \\
& = 2y
\end{aligned}
$ 
<br><br>
**Step 2. expecation values** <br>
So the expectation values of $X$ and $Y$ are <br>
$
\begin{aligned}
E[X] & = \int^1_0 x f_X(x) dx \\
& \int^1_0 x(2-2x) dx \\
& = \left[ x^2 - \frac{2}{3} x^3 \right]^1_0 \\
& = \frac{1}{3}
\end{aligned}
$ 
<br>
$
\begin{aligned}
E[Y] & = \int^1_0 y f_Y(y) dx \\
& \int^1_0 y \cdot 2y dy \\
& = \left[ \frac{2}{3} y^3 \right]^1_0 \\
& = \frac{2}{3} \\
\end{aligned}
$
<br>
$
\begin{aligned}
E[XY] & = \int^1_0 \int^1_x xy f(x, y) dy dx \\
& = \int^1_0 \int^1_x 2xy dy dx \\
& = \int^1_0 \left[ x y^2 \right]^1_x dx \\
& = \int^1_0 x - x^3 dx \\
& = \left[ \frac{1}{2}x^2 - \frac{1}{4}x^4\right]^1_0 \\
& = \frac{1}{4}
\end{aligned}
$

**Step 3. Variance**<br>
Then, we calculate the variance of the variables.<br><br>
$
\begin{aligned}
\sigma_X^2 & = E[(X - E[X])^2] \\
& = E[X^2] - E[X] ^2 \\
& = \int^1_0 x^2 f_X(x) dx - E[X] ^2 \\
& = \int^1_0 x^2 (2-2x) dx - E[X] ^2 \\
& = \left[ \frac{2}{3}x^3 - \frac{1}{2} x^4 \right]^1_0 - E[X] ^2 \\
& = \frac{1}{6} - \frac{1}{9} \\
& = \frac{1}{18}
\end{aligned}
$
<br>
$
\begin{aligned}
\sigma_Y^2 & = E[(Y - E[Y])^2] \\
& = E[Y^2] - E[Y] ^2 \\
& = \int^1_0 y^2 f_Y(y) dy - E[Y] ^2 \\
& = \int^1_0 2 y^3 dy - E[Y] ^2 \\
& = \left[ \frac{1}{2}x^4 \right]^1_0 - E[Y] ^2\\
& = \frac{1}{2} - \frac{4}{9} \\
& = \frac{1}{18}
\end{aligned}
$

**Step 4.Covariance**<br>
Then, the covariance is<br><br>
$
\begin{aligned}
Cov(X, Y) & = E[XY] - E[X]E[Y] \\
& = \frac{1}{4} - \frac{1}{3} \cdot \frac{2}{3} \\
& = \frac{1}{36}
\end{aligned}
$

**Step 5.The correlation coefficient**<br>
Finally, we can find the correlation coeffinient<br><br>
$
\begin{aligned}
\rho_{XY} & =\frac{Cov(X,Y)}{\sigma_X \sigma_Y} \\
& = \frac{1}{36} / \sqrt{\frac{1}{18}} \sqrt{\frac{1}{18}} \\
& = \frac{1}{36} \cdot 18 \\
& = \frac{1}{2}
\end{aligned}
$

**Problem 5.**
If a random variables $X$ is defined such that
$$ E[(X-1)^2] = 10\  \mathrm{and}\  E[(X-2)^2] = 6$$
find $\mu$ and $\sigma^2$.

----

**[Sol]**<br>
The first equation can be expanded to<br>
$E[(X-1)^2] = E[X^2] -2E[X] + 1 = 10$ $\cdots (1)$<br>
And also, the second equation can be expanded to<br>
$E[(X-2)^2] = E[X^2] -4E[X] + 4 = 6$ $\cdots (2)$<br>
Using the relation of the expectation value with means and variance, we can calculate that<br><br>
$
\begin{aligned}
(1) & =  E[X^2] -E[X] - E[X]  + 1 \\
& = \sigma^2 - \mu + 1  = 10 \\
\\
(2) & = E[X^2] -E[X] - 3E[X]  + 4 \\
& = \sigma^2 - 3\mu + 4  = 6
\end{aligned}
$
<br>
Solving the two equation,<br><br>
$
\begin{cases}
\sigma^2 - \mu + 1  = 10 \\
\sigma^2 - 3\mu + 4  = 6
\end{cases}
$
<br>
Then,<br><br>
$
\begin{cases}
\mu = \frac{7}{2} \\
\sigma^2 = \frac{25}{2}
\end{cases}
$

**Problem 6.**<br>
(a) (Markov's Inequality) Show that if $X$ is a nonnegative random variable and $a>0$, then
$$ P(X \geq a) \leq \frac{E[X]}{a}$$
(b) (Chebyshev's Inequality) Show that
$$ P(|X - E[X]| \geq a) \leq \frac{Var(X)}{a^2}$$
----

**[Sol.]**<br>
(a) From the expectation value of $X$ <br><br>
$
\begin{aligned}
E[X] & = \int^\infin_{-\infin} xf(x) \ dx \\
& = \int^\infin_0 xf(x) \ dx
\end{aligned}
$
<br>
From this, we can derive that<br><br>
$
\begin{aligned}
E[X] & = \int^\infin_0 xf(x) \ dx \\
& = \int^a_0 xf(x) \ dx + \int^\infin_a xf(x) \ dx \geq \int^\infin_a xf(x) \ dx \geq \int^\infin_a af(x) \\
& = a \int^\infin_a f(x) = a P(X>a)
\end{aligned}
$
<br>
(b) From Markov's inequlity, we can modify the statement<br><br>
$
P(|X - \mu| \leq a) = P((X - \mu)^2 \leq a^2) \geq \frac{E[(X - \mu)^2]}{a^2} = \frac{Var(X)}{a^2}
$
