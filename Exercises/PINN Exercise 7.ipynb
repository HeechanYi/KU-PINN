**Problem 1.**
Prove that
$$P(A^C \cap B^C) = 1 + P(A \cap B) - P(A) -P(C) $$
----

**[Sol]**<br>
By de Morgan's law,<br>
$$(A \cap B)^C  = A^C \cup B^C$$

$
\begin{aligned}
\text{then} \, P(A^C \cap B^C) & =  P((A \cup B)^C)\\
 & = 1 - P(A \cup B) \quad \text{where} \quad P(A \cup B) = P(A) + P(B) - P(A\cap B)\\
 & = 1 - (P(A) + P(B) - P(A\cap B))\\
 & = 1 + P(A\cap B) - P(A) - P(B) \ \ \ \ _\Box
\end{aligned}
$

**Problem 2.** In a certain region of the country it is known from past experience that the probability of selecting an adult over 40 years of age with cancer is 0.05. If the probability of a doctor correctly diagnosing a person with cancer as having the disease is 0.78 and the probability of incorrectly diagnosing a person without cancer as having the disease is 0.06, **what is the probability that an adult over 40 years of age is diagnosed as having cancer?**

----

**[Sol]**   
From the problem, we could define that<br/>  
$P(\text{selecting an adult over 40 years of age with cancer}) := P(C) = 0.05$<br/> 
$P(\text{correct diagnosing with cancer}) := P(p \cap C) = 0.78$<br/>
$P(\text{incorrect diagnosing without cancer}) := P(p \cap C^C) = 0.06$<br/><br/>
Then, the problem is asking about $P(\text{Diagnose as having cancer}) = P(p)$<br/><br/><br/>

$
\begin{aligned}
P(p) & =  P(p|C) + P(p|C^C) \\
 & = P(p \cap C)P(C) + P(p \cap C^C)P(C^C) \\
 & = 0.78*0.05 + 0.06*0.95 = 0.096
\end{aligned}
$

**Problem 3.** Consider the density function
$$
f(x) = 
\begin{cases}
k\sqrt{x} & \quad & 0<x<1 \\
0, & \quad & \textit{otherwise} 
\end{cases}
$$
(a) Evaluate $k$.<br>
(b) Find $P(0.3 < X < 0.6)$.

----

**[Sol]**<br>
(a) To find $k$, we use the fact that the total integral of the density function over its entire range must equal 1.

$$
\int_{-\infty}^{\infty} f(x) \, dx = 1
$$

Given the density function:

$$
\int_{0}^{1} k\sqrt{x} \, dx = 1
$$

We know that:

$$
\int \sqrt{x} \, dx = \frac{2}{3} x^{3/2}
$$

So:

$$
k \left[ \frac{2}{3} x^{3/2} \right]_{0}^{1} = 1
$$

Evaluating the definite integral:

$
\begin{aligned}
&k \left( \frac{2}{3} (1)^{3/2} - \frac{2}{3} (0)^{3/2} \right) = 1 \Rightarrow k \left( \frac{2}{3} \right) = 1 \Rightarrow k = \frac{3}{2}
\end{aligned}
$

(b) To find $P(0.3 < X < 0.6)$, we integrate the density function from 0.3 to 0.6:

$$
P(0.3 < X < 0.6) = \int_{0.3}^{0.6} f(x) \, dx
$$

Substituting $f(x)$:

$$
P(0.3 < X < 0.6) = \int_{0.3}^{0.6} \frac{3}{2} \sqrt{x} \, dx
$$

Solving the integral:

$$
\frac{3}{2} \int_{0.3}^{0.6} \sqrt{x} \, dx
$$

Using the antiderivative of $\sqrt{x}$:

$$
\frac{3}{2} \left[ \frac{2}{3} x^{3/2} \right]_{0.3}^{0.6}
$$

Evaluating the definite integral:

$
\begin{aligned}
&\frac{3}{2} \left( \frac{2}{3} (0.6)^{3/2} - \frac{2}{3} (0.3)^{3/2} \right) = \left( (0.6)^{3/2} - (0.3)^{3/2} \right)\\
&= \left( \sqrt{0.6^3} - \sqrt{0.3^3} \right)\\
&= \left( \sqrt{0.216} - \sqrt{0.027} \right)\\
&= (0.464 - 0.164) = 0.3
\end{aligned}
$

So, $P(0.3 < X < 0.6) = 0.3$.

**Problem 4.** Let $X$ and $Y$ denote the lengths of life, in years, of two components in an electonic system. If the joint density function of these variables is
$$
f(x,y) = 
\begin{cases}
e^{-(x+y)} & x > 0, y > 0 \\
0, & \textit{elsewhere} 
\end{cases}
$$
find $P(0<X<1 | Y = 2)$

---

**[Sol]**<br>
To find the conditional probability $P(0 < X < 1 | Y = 2)$, we use the definition of conditional probability density function:

$$
P(0 < X < 1 | Y = 2) = \frac{P(0 < X < 1 \cap Y = 2)}{P(Y = 2)}
$$

First, we find $P(Y = 2)$. The marginal density function of $Y$ is obtained by integrating the joint density function over all possible values of $X$:

$$
f_Y(y) = \int_{0}^{\infty} f(x,y) \, dx
$$

Given $f(x,y) = e^{-(x+y)}$ for $x > 0$ and $y > 0$:

$
\begin{aligned}
f_Y(2) & =\int_{0}^{\infty} e^{-(x+2)} \, dx \\
& = e^{-2} \int_{0}^{\infty} e^{-x} \, dx \\
& = e^{-2} \left[ -e^{-x} \right]_{0}^{\infty}\\
& = e^{-2} \left( 0 - (-1) \right) = e^{-2}
\end{aligned}
$

Next, we find $P(0 < X < 1 \cap Y = 2)$, which is the integral of the joint density function over the given range:

$
\begin{aligned}
P(0 < X < 1 \cap Y = 2) &= \int_{0}^{1} f(x,2) \, dx \\
&= \int_{0}^{1} e^{-(x+2)} \, dx \\
&= e^{-2} \int_{0}^{1} e^{-x} \, dx \\
&= e^{-2} \left[ -e^{-x} \right]_{0}^{1} \\
&= e^{-2} \left( -e^{-1} - (-1) \right) = e^{-2} \left( 1 - e^{-1} \right)
\end{aligned}
$

Finally, we find the conditional probability:

$
\begin{aligned}
P(0 < X < 1 | Y = 2) &= \frac{P(0 < X < 1 \cap Y = 2)}{P(Y = 2)} = \frac{e^{-2} \left( 1 - e^{-1} \right)}{e^{-2}} \\
&= 1 - e^{-1}
\end{aligned}
$

So, $P(0 < X < 1 | Y = 2) = 1 - e^{-1}$.

**Problem 5.** The joint density function of the random varaibles $X$ and $Y$ is
$$
f(x, y) =
\begin{cases}
6x & 0 < x < 1, 0, 0 < y < 1- x \\
0, & \textit{elsewhere} 
\end{cases}
$$
(a) Show that $X$ and $Y$ are not independent.<br>
(b) Find $P(X>0.3|Y=0.5)$

----

**[Sol]**<br>
(a) To show that $X$ and $Y$ are not independent, we need to check if the joint density function $f(x, y)$ can be written as the product of the marginal density functions $f_X(x)$ and $f_Y(y)$.

First, we find the marginal density function of $X$:

$$
f_X(x) = \int_{0}^{1-x} f(x, y) \, dy
$$

Given $f(x, y) = 6x$ for $0 < x < 1$ and $0 < y < 1 - x$:

$$
f_X(x) = \int_{0}^{1-x} 6x \, dy = 6x \int_{0}^{1-x} \, dy = 6x (1 - x)
$$

Next, we find the marginal density function of $Y$:

$$
f_Y(y) = \int_{0}^{1-y} f(x, y) \, dx
$$

Given $f(x, y) = 6x$ for $0 < x < 1$ and $0 < y < 1 - x$:

$$
f_Y(y) = \int_{0}^{1-y} 6x \, dx = 6 \int_{0}^{1-y} x \, dx = 6 \left[ \frac{x^2}{2} \right]_{0}^{1-y} = 6 \left( \frac{(1-y)^2}{2} \right) = 3(1-y)^2
$$

Now, we check if $f(x, y) = f_X(x) f_Y(y)$:

$$
f_X(x) f_Y(y) = 6x (1 - x) \cdot 3(1 - y)^2 = 18x (1 - x)(1 - y)^2
$$

Since $f(x, y) = 6x \neq 18x (1 - x)(1 - y)^2$, $X$ and $Y$ are not independent.

(b) To find $P(X > 0.3 | Y = 0.5)$, we use the definition of conditional probability density function:

$$
P(X > 0.3 | Y = 0.5) = \frac{P(X > 0.3 \cap Y = 0.5)}{P(Y = 0.5)}
$$

First, we find $P(Y = 0.5)$. The marginal density function of $Y$ is:

$$
f_Y(0.5) = 3(1 - 0.5)^2 = 3(0.5)^2 = 3 \cdot 0.25 = 0.75
$$

Next, we find $P(X > 0.3 \cap Y = 0.5)$, which is the integral of the joint density function over the given range:

$$
P(X > 0.3 \cap Y = 0.5) = \int_{0.3}^{0.5} f(x, 0.5) \, dx
$$

Given $f(x, y) = 6x$ for $0 < x < 1$ and $0 < y < 1 - x$:

$$
P(X > 0.3 \cap Y = 0.5) = \int_{0.3}^{0.5} 6x \, dx = 6 \int_{0.3}^{0.5} x \, dx = 6 \left[ \frac{x^2}{2} \right]_{0.3}^{0.5} = 6 \left( \frac{0.5^2}{2} - \frac{0.3^2}{2} \right) = 6 \left( \frac{0.25}{2} - \frac{0.09}{2} \right) = 6 \left( 0.125 - 0.045 \right) = 6 \cdot 0.08 = 0.48

**Problem 6.** A chemical system that results from a chemical reaction has two important components among others in a blend. The joint distribution describing the proportions $X_1$ and $X_2$ of these two components is given by
$$
f(x_1, x_2) = 
\begin{cases}
2, & 0 < x_1 < x_2 <1 \\
0, & \textit{elsewhere}
\end{cases}
$$
(a) Give the marginal distribution of $X_1$ and $X_2$<br>
(b) What is the probability that component propotions produce the results $X_1 < 0.2$ and $X_2 > 0.5$?<br>
(c) Give the conditional distribution $f(x_1|x_2)$

-----

**[Sol]**<br>
(a) To find the marginal distribution of $X_1$, we integrate the joint density function over all possible values of $X_2$:

$$
f_{X_1}(x_1) = \int_{x_1}^{1} f(x_1, x_2) \, dx_2
$$

Given $f(x_1, x_2) = 2$ for $0 < x_1 < x_2 < 1$:

$$
f_{X_1}(x_1) = \int_{x_1}^{1} 2 \, dx_2 = 2 \int_{x_1}^{1} \, dx_2 = 2 (1 - x_1)
$$

To find the marginal distribution of $X_2$, we integrate the joint density function over all possible values of $X_1$:

$$
f_{X_2}(x_2) = \int_{0}^{x_2} f(x_1, x_2) \, dx_1
$$

Given $f(x_1, x_2) = 2$ for $0 < x_1 < x_2 < 1$:

$$
f_{X_2}(x_2) = \int_{0}^{x_2} 2 \, dx_1 = 2 \int_{0}^{x_2} \, dx_1 = 2 x_2
$$

(b) To find the probability that $X_1 < 0.2$ and $X_2 > 0.5$, we integrate the joint density function over the given range:

$$
P(X_1 < 0.2 \cap X_2 > 0.5) = \int_{0.5}^{1} \int_{0}^{0.2} f(x_1, x_2) \, dx_1 \, dx_2
$$

Given $f(x_1, x_2) = 2$ for $0 < x_1 < x_2 < 1$:

$$
P(X_1 < 0.2 \cap X_2 > 0.5) = \int_{0.5}^{1} \int_{0}^{0.2} 2 \, dx_1 \, dx_2 = 2 \int_{0.5}^{1} \int_{0}^{0.2} \, dx_1 \, dx_2
$$

Evaluating the integral:

$$
2 \int_{0.5}^{1} \left[ x_1 \right]_{0}^{0.2} \, dx_2 = 2 \int_{0.5}^{1} 0.2 \, dx_2 = 2 \cdot 0.2 \int_{0.5}^{1} \, dx_2 = 2 \cdot 0.2 (1 - 0.5) = 2 \cdot 0.2 \cdot 0.5 = 0.2
$$

So, $P(X_1 < 0.2 \cap X_2 > 0.5) = 0.2$.

(c) To find the conditional distribution $f(x_1|x_2)$, we use the definition of conditional probability density function:

$$
f(x_1|x_2) = \frac{f(x_1, x_2)}{f_{X_2}(x_2)}
$$

Given $f(x_1, x_2) = 2$ for $0 < x_1 < x_2 < 1$ and $f_{X_2}(x_2) = 2 x_2$:

$$
f(x_1|x_2) = \frac{2}{2 x_2} = \frac{1}{x_2} \quad \text{for} \quad 0 < x_1 < x_2 < 1
$$

**Problem 7.**<br>
(a) Use Python to approximate the mean and variance of $Y = 2X + 1$ and $Z = X^2$ where $X \sim \mathcal{N}(0,1)$.<br>
(b) Plot the histogram of $Y$ and $Z$ with 10,000 samples.<br>
(c) Plot the probability functions of $\mathcal{N}(1,4)$ and $\chi(1)$(chi-squared distribution) and compare them with the histograms of $Y$ and $Z$, respectively.

----

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
````

````python
# Pb 7.(a)
# Generate 10,000 samples from the normal distribution N(0,1)
X = np.random.normal(0, 1, 10000)

# Compute Y and Z
Y = 2 * X + 1
Z = X ** 2

# Approximate the mean and variance of Y and Z
mean_Y = np.mean(Y)
variance_Y = np.var(Y)
mean_Z = np.mean(Z)
variance_Z = np.var(Z)
````

````python
# Pb 7.(b)
# Plot the histogram
plt.figure(figsize=(12, 6))

# Plot histogram of Y
plt.subplot(1, 2, 1)
plt.hist(Y, bins=50, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram of Y')
plt.xlabel('Y')
plt.ylabel('Frequency')
plt.grid(True)

# Plot histogram of Z
plt.subplot(1, 2, 2)
plt.hist(Z, bins=50, alpha=0.7, color='green', edgecolor='black')
plt.title('Histogram of Z')
plt.xlabel('Z')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
````

````python
# Pb 7.(c)
# Make the theoretical distribution of normal distribution and chi-square distribution
def normal(mean, variance, x):
    return 1 / np.sqrt(2 * np.pi * variance) * np.exp(- (x - mean) ** 2 / (2 * variance))

def chisquare(df, x):
    return 1 / (2 ** (df / 2) * np.math.gamma(df / 2)) * x ** (df / 2 - 1) * np.exp(-x / 2)


xy = np.linspace(-10, 10, 1000)
xz = np.linspace(.001, 10, 1000)
y = normal(1, 4, xy)
z = chisquare(1, xz)

# Plot the theoretical distribution and the histogram of Y and Z
plt.figure(figsize=(12, 6))

# Plot the theoretical distribution of Y and the histogram of Y
plt.subplot(1, 2, 1)
plt.plot(xy, y, color='red', label='Theoretical distribution of Y')
plt.hist(Y, bins=50, alpha=0.7, density=True, color='blue', edgecolor='black', label='Histogram of Y')
plt.title('Histogram of Y')
plt.xlabel('Y / x')
plt.ylabel('Frequency')
plt.grid(True)

# Plot the theoretical distribution of Z and the histogram of Z
plt.subplot(1, 2, 2)
plt.plot(xz, z, color='red', label='Theoretical distribution of Z')
plt.hist(Z, bins=50, alpha=0.7, density=True, color='green', edgecolor='black', label='Histogram of Z')
plt.title('Histogram of Z')
plt.xlabel('Z / x')
plt.ylabel('Frequency ')
plt.grid(True)

plt.tight_layout()
plt.show()
````
