**Definition**
 $\textit{We say a random variable X has a chi-squared distribution with degree of freedom ν if its p.d.f. is given by}$
$$
\chi^2 (x;\nu) =
\begin{cases}
\frac{x^{\nu/2 - 1} e^{-x / 2}}{2^{\nu/2} \Gamma(\nu/2)}, & x > 0 \\
0, & otherwise.
\end{cases}
$$

**Problem 1.**<br>
Let $X \sim \mathcal{N}(0, 1)$. Show that $Y = X^2$ has a chi-squared distribution with 1 degree of freedom.

----

**[Sol.]**<br>
Before solving the problem, we have to clarify the mean of 'same distribution'.<br>
To say that two vairables $X$ and $Y$ have a same distribution, they have to have the same 'moment generating function',
$$E[e^{tY}] = E[e^{tX}]$$
Then, calculating the moment generating function of $Y$<br><br>
$
\begin{aligned}
E[e^{tY}] = E[e^{tX^2}] & = \int^\infty_{-\infty} e^{tx^2} \frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}} dx \\
& = \int^\infty_{-\infty} \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2} x^2 (1 - 2t)}
\end{aligned}
$
<br>
And also, calculating the generating function of $\chi^2$ with 1 degree of freedom<br><br>
$
\begin{aligned}
&\int^\infty_{0} e^{tx} \frac{x^{-1/2} e^{-x/2}}{2^{1/2} \cdot \Gamma(\frac{1}{2})}dx \\
&= \int^\infty_{0} \frac{1}{\sqrt{2\pi}} x^{-1/2} e^{-\frac{1}{2}x(1 - 2t)} dx
\end{aligned}
$
<br>
Then substituete $x = y^2$ so that $dx = 2|y|dy$, which is $dy = \frac{1}{2}|y|^{-1/2} dx$ and the range can be expanded $[0, \infty] \to [-\infty, \infty]$ <br><br>
$
\begin{aligned}
&\int^\infty_{0} \frac{1}{\sqrt{2\pi}} x^{-1/2} e^{-\frac{1}{2}x(1 - 2t)} dx \\
& = \int^\infty_{0} \frac{1}{\sqrt{2\pi}} (y^2)^{-1/2} e^{-\frac{1}{2}y^2(1 - 2t)} 2|y|dy\\
& = 2 \int^\infty_{0} \frac{1}{\sqrt{2\pi}} |y|^{-1} e^{-\frac{1}{2}y^2(1 - 2t)} |y|dy \\
& = \int^\infty_{0} \frac{1}{\sqrt{2\pi}} |y|^{-1} e^{-\frac{1}{2}y^2(1 - 2t)} |y|dy + \int^\infty_{0} \frac{1}{\sqrt{2\pi}} |y|^{-1} e^{-\frac{1}{2}y^2(1 - 2t)} |y|dy \quad \cdots \  \cdots (*)\\
& = \int^\infty_{0} \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2}y^2(1 - 2t)} dy - \int^{-\infty}_{0} \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2}y^2(1 - 2t)} dy \\
& = \int^\infty_{0} \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2}y^2(1 - 2t)} dy + \int^{0}_{-\infty} \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2}y^2(1 - 2t)} dy \\
& = \int^\infty_{-\infty} \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2}y^2(1 - 2t)} dy
\end{aligned}
$
<br>
Beyond step $(*)$, We have to carefully consider that the range of $y$ can be in the negative number. Therefore, the next equation's first term notes the integral in the positive number and the second term means the integral in the negative number.

**Problem 2.**<br>
Let $X$ have the probability distribution
$$
f(x) = 
\begin{cases}
1, & 0< x <1 \\
0, & \mathrm{elsewhere} \\
\end{cases}
$$
Show that the random variable $Y = -2\ln X$ has a chi-squared distribution with 2 degrees of freedom.

----

**[Sol.]**<br>
Calculating the generating function of $Y = -2\ln X$,<br>
$$ E[e^{tY}] = E[e^{-2t\ln X}] = E[X^{-2t}] = \int^1_0 x^{-2t} dx  $$
Then, finding the generating function of $\chi^2$ with 2 degrees of freedom. <br><br>
$
\begin{aligned}
& \int^\infty_{0} e^{tx} \frac{e^{-x/2}}{2^{1} \cdot \Gamma(1)}dx \\
& = \int^\infty_{0} \frac{1}{2} e^{-\frac{1}{2}x(1-2t)}dx
\end{aligned}
$
<br>
Substituting the variable<br><br>
$
\begin{aligned}
& \int^\infty_{0} \frac{1}{2} e^{-\frac{1}{2}x(2-t)}dx\\
& =  - \int^{-\infty}_{0} e^{a(1-2t)}da \\
& =  \int^{0}_{-\infty} e^{a(1-2t)} da \\
& =  \int^{1}_{0} e^{(1-2t)\ln b} b^{-1} db \\
& =  \int^{1}_{0} b^{(1-2t)} b^{-1} db \\
& =  \int^{1}_{0} b^{-2t}  db
\end{aligned}
$

**Problem 3.**<br>
A random variable $X$ has the Poisson distribution $p(x;\mu) = \frac{\mu^x e^{-\mu}}{x!}$ for $ x = 0, 1, 2, ...$. Show that the moment generating function of $X$ is
$$ M_X(t) = e^{\mu(e^t - 1)}$$
Using $M_X(t)$, find the mean and variance of the Poisson distribution.

----

**[Sol.]**<br>
Using the relationship of the exponential and sum:
$$ \sum^\infty_{x=0} \frac{a^x}{x!} = e^a$$
The moment generating function will be<br><br>
$
\begin{aligned}
M_X(t) & = E[e^{tX}] \\
& = \sum^\infty_{x = 0} e^{tx} \frac{\mu^x e^{-\mu}}{x!} \\
& = e^{-\mu} \sum^\infty_{x = 0} e^{tx} \frac{\mu^x}{x!} \\
& = e^{-\mu} \sum^\infty_{x = 0}  \frac{(e^{t}\mu)^x}{x!} \\
& = e^{-\mu} e^{e^{t}\mu} \\
& = e^{\mu(e^t - 1)} 
\end{aligned}
$

**Problem 4.(Python)**<br>
Write a Python program that calculates the sample mean and sample variance of given random samples. Use this program to test whether these estimates converge to their true values as $n → ∞$, specifically for uniform and normal distributions.

----

```python
import numpy as np
import matplotlib.pyplot as plt

from tqdm import tqdm
```

```python
def mean(samples):
    return sum(samples) / len(samples)

def variance(samples):
    return sum((x - mean(samples))**2 for x in samples) / len(samples)
```

```python
### Sample size for large number of samples
n = np.logspace(1, 5, num = 10, dtype = int)

### True mean and variance
true_mean_uniform, true_variance_uniform = 0.5, 1/12    # Uniform distribution(Mean, Variance)
true_mean_gaussian, true_variance_gaussian = 0, 1       # Gaussian distribution(Mean, Variance)

### Sample mean and variance
sample_mean_uniform, sample_variance_uniform = [], []
sample_mean_gaussian, sample_variance_gaussian = [], []

### Generate samples
for i in tqdm(n):
    ### Uniform distribution
    samples_uniform = np.random.uniform(0, 1, i)
    sample_mean_uniform.append(mean(samples_uniform))
    sample_variance_uniform.append(variance(samples_uniform))

    ### Gaussian distribution
    samples_gaussian = np.random.normal(0, 1, i)
    sample_mean_gaussian.append(mean(samples_gaussian))
    sample_variance_gaussian.append(variance(samples_gaussian))
```

```python
### Plotting Convergence
fig, axis = plt.subplots(2, 2, figsize = (15, 11))

### Uniform distribution
axis[0, 0].plot(n, sample_mean_uniform, label = 'Sample Mean')
axis[0, 0].axhline(y = true_mean_uniform, color = 'r', linestyle = '--', label = 'True Mean')
axis[0, 0].set_xscale('log')
axis[0, 0].set_xlabel('Sample Size')
axis[0, 0].set_ylabel('Mean')
axis[0, 0].legend()
axis[0, 0].grid(True)
axis[0, 0].set_title('Uniform Distribution')

axis[1, 0].plot(n, sample_variance_uniform, label = 'Sample Variance')
axis[1, 0].axhline(y = true_variance_uniform, color = 'r', linestyle = '--', label = 'True Variance')
axis[1, 0].set_xscale('log')
axis[1, 0].set_xlabel('Sample Size')
axis[1, 0].set_ylabel('Variance')
axis[1, 0].legend()
axis[1, 0].grid(True)
axis[1, 0].set_title('Uniform Distribution')

### Gaussian distribution
axis[0, 1].plot(n, sample_mean_gaussian, label = 'Sample Mean')
axis[0, 1].axhline(y = true_mean_gaussian, color = 'r', linestyle = '--', label = 'True Mean')
axis[0, 1].set_xscale('log')
axis[0, 1].set_xlabel('Sample Size')
axis[0, 1].set_ylabel('Mean')
axis[0, 1].legend()
axis[0, 1].grid(True)
axis[0, 1].set_title('Gaussian Distribution')

axis[1, 1].plot(n, sample_variance_gaussian, label = 'Sample Variance')
axis[1, 1].axhline(y = true_variance_gaussian, color = 'r', linestyle = '--', label = 'True Variance')
axis[1, 1].set_xscale('log')
axis[1, 1].set_xlabel('Sample Size')
axis[1, 1].set_ylabel('Variance')
axis[1, 1].legend()
axis[1, 1].grid(True)
axis[1, 1].set_title('Gaussian Distribution')
```

**Problem 5.(Python)**<br>
Write a Python program to verify the results of the Central Limit Theorem.<br>
(Hint : The Central Limit Theorem is about the *distribution* of the smaple mean $\bar{X}_n$)

----

```python
### Central Limit Theorem
def apply_CLT(distribution_data, sample_size, total_samples):
  sample_mean = []
  for _ in range(total_samples):                                        #Iterating over total samples(n)
    sample = np.random.choice(distribution_data, size = sample_size)    #Randomly choosing samples from the distribution data
    mean = np.mean(sample)                                              #Finding mean of random chose samples using numpy
    sample_mean.append(mean)     
     
  return sample_mean
```

```python
### Making Distrubition Data
gaussian_data = np.random.normal(0, 1, 10000)    #Gaussian Distribution Data
uniform_data = np.random.uniform(0, 1, 10000)    #Uniform Distribution Data
expone_data = np.random.exponential(1, 10000)    #Exponential Distribution Data

### Applying CLT
sample_size = [10, 20, 30, 40, 50]    #Sample Size
total_samples = 1000                  #Total Samples

sample_mean_gaussian = [apply_CLT(gaussian_data, i, total_samples) for i in sample_size]    #Applying CLT on Gaussian Data
sample_mean_uniform = [apply_CLT(uniform_data, i, total_samples) for i in sample_size]      #Applying CLT on Uniform Data
sample_mean_expone = [apply_CLT(expone_data, i, total_samples) for i in sample_size]        #Applying CLT on Exponential Data

### Plotting CLT
fig, axis = plt.subplots(3, 1, figsize = (8, 15))

### Gaussian Distribution
for i in range(len(sample_size)):
    axis[0].hist(sample_mean_gaussian[i], bins = 50, alpha = 0.5, density=True, label = f'Sample Size: {sample_size[i]}')
    axis[0].set_xlabel('Sample Mean')
    axis[0].set_ylabel('Frequency')
    axis[0].legend()
    axis[0].grid(True)
    axis[0].set_title('Gaussian Distribution')

### Uniform Distribution
for i in range(len(sample_size)):
    axis[1].hist(sample_mean_uniform[i], bins = 50, alpha = 0.5, density=True, label = f'Sample Size: {sample_size[i]}')
    axis[1].set_xlabel('Sample Mean')
    axis[1].set_ylabel('Frequency')
    axis[1].legend()
    axis[1].grid(True)
    axis[1].set_title('Uniform Distribution')

### Exponential Distribution
for i in range(len(sample_size)):
    axis[2].hist(sample_mean_expone[i], bins = 50, alpha = 0.5, density=True, label = f'Sample Size: {sample_size[i]}')
    axis[2].set_xlabel('Sample Mean')
    axis[2].set_ylabel('Frequency')
    axis[2].legend()
    axis[2].grid(True)
    axis[2].set_title('Exponential Distribution')

plt.show()
```
